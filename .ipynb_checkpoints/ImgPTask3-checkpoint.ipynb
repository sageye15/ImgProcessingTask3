```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 3]
```

# Task 1

```python
img = cv2.imread('./Task1-assets/1.jpeg')
img = cv2.GaussianBlur(img, (3,3),100)

grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(grey, 2, 70)
plt.imshow(canny)
plt.show()

lines = cv2.HoughLines(canny, 1, np.pi/180, 10)
for line in lines:
    print(line)
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

plt.imshow(img)
plt.show()
```
